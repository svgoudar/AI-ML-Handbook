

## Type I and Type II Errors in Hypothesis Testing

### Key Concepts

1. **Reality vs Decision**

   * **Reality:** The null hypothesis (H₀) is either **true** or **false**.
   * **Decision:** Based on hypothesis testing, we **accept/reject H₀**.

2. **Possible Outcomes**

| Outcome | Decision         | Reality | Correct/Error | Type of Error |
| ------- | ---------------- | ------- | ------------- | ------------- |
| 1       | Reject H₀        | False   | Correct       | —             |
| 2       | Reject H₀        | True    | Error         | **Type I**    |
| 3       | Accept/Retain H₀ | False   | Error         | **Type II**   |
| 4       | Accept/Retain H₀ | True    | Correct       | —             |

---

### Explanation*

* **Type I Error (α):** Rejecting H₀ when it is actually **true**.

  * “False positive” in testing terms.
* **Type II Error (β):** Accepting H₀ when it is actually **false**.

  * “False negative” in testing terms.

---

### Key Points

* Type I and Type II errors represent the **risk of incorrect decisions** in hypothesis testing.
* Understanding these errors is important in statistics and also in **machine learning** (confusion matrix, error rates).
* Correct decisions occur when we **reject a false null** or **accept a true null**.



## Example 1: Medical Testing

* **Scenario:** A doctor tests for a disease.

  * **H₀ (Null Hypothesis):** Patient does **not** have the disease.
  * **H₁ (Alternative Hypothesis):** Patient **has** the disease.

| Error Type  | What Happens                                                                                        |
| ----------- | --------------------------------------------------------------------------------------------------- |
| **Type I**  | Doctor diagnoses the patient as **diseased**, but in reality they are **healthy**. (False positive) |
| **Type II** | Doctor diagnoses the patient as **healthy**, but in reality they are **diseased**. (False negative) |

---

## Example 2: Quality Control in Manufacturing

* **Scenario:** A factory checks if a batch of products meets quality standards.

  * **H₀:** The batch **meets** quality standards.
  * **H₁:** The batch **does not meet** quality standards.

| Error Type  | What Happens                                                                                      |
| ----------- | ------------------------------------------------------------------------------------------------- |
| **Type I**  | Reject a batch that **actually meets** quality standards. (Unnecessary rejection → loss of money) |
| **Type II** | Accept a batch that **does not meet** quality standards. (Defective products reach customers)     |

---

## Example 3: Exam Cheating Detection

* **Scenario:** Detecting if a student cheated.

  * **H₀:** Student **did not cheat**.
  * **H₁:** Student **cheated**.

| Error Type  | What Happens                                                |
| ----------- | ----------------------------------------------------------- |
| **Type I**  | Accuse a student of cheating who **actually didn’t cheat**. |
| **Type II** | Let a student go free who **actually cheated**.             |

---

### Key Takeaways from Examples

* **Type I Error (α):** Rejecting H₀ when it is actually true → **false alarm**.
* **Type II Error (β):** Accepting H₀ when it is actually false → **missed detection**.
* The **balance** between Type I and Type II errors depends on the **consequence of errors** in real-life situations.
