## Pearson Correlation Coefficient

### Definition

The **Pearson correlation coefficient (r)** measures the **strength** and **direction** of the **linear relationship** between two continuous variables.

It tells us:

* *Are X and Y related?*
* *If yes, how strong is the relationship?*
* *And is it positive or negative?*

---

### Formula

For a sample:

$$
r = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2 \sum_{i=1}^n (y_i - \bar{y})^2}}
$$

Where:

* $x_i, y_i$ = individual data points
* $\bar{x}, \bar{y}$ = sample means
* $n$ = number of data points

---

### Range

$$
-1 \leq r \leq +1
$$

* **+1** → Perfect positive linear correlation (as X ↑, Y ↑ exactly on a straight line).
* **-1** → Perfect negative linear correlation (as X ↑, Y ↓ exactly on a straight line).
* **0** → No linear correlation (but there could still be a nonlinear relationship).

---

### Interpretation of r

* **0.9 to 1.0** → Very strong positive
* **0.7 to 0.9** → Strong positive
* **0.5 to 0.7** → Moderate positive
* **0.3 to 0.5** → Weak positive
* **0.0 to 0.3** → Very weak/negligible

(Same applies with negative signs for negative correlation.)

---

### Example

Suppose we study **hours studied (X)** vs **exam score (Y)**:

| Hours Studied (X) | Score (Y) |
| ----------------- | --------- |
| 2                 | 50        |
| 4                 | 65        |
| 6                 | 80        |
| 8                 | 90        |
| 10                | 95        |

When we compute $r$, it comes out close to **+0.98** → almost a perfect positive linear relationship.

---

### Properties

1. **Unitless measure** → standardizes covariance.
2. **Symmetric** → $r(X, Y) = r(Y, X)$.
3. **Only measures linear relationships** (not nonlinear).
4. **Sensitive to outliers** → one extreme point can distort r heavily.

---

### When to Use

* Checking **feature-target relationships** in ML.
* Identifying **collinearity** among predictors.
* Financial analysis (e.g., correlation between two stock returns).
* Health sciences (e.g., exercise vs cholesterol levels).

---

**In short**:
Pearson’s r = standardized covariance, bounded between \[-1, +1], and it’s the go-to metric for measuring linear associations.


````python
import numpy as np
import matplotlib.pyplot as plt

# Example data from the table
hours = np.array([2, 4, 6, 8, 10])
scores = np.array([50, 65, 80, 90, 95])

# Scatter plot
plt.figure(figsize=(6,4))
plt.scatter(hours, scores, color='teal', s=80)
plt.plot(hours, np.poly1d(np.polyfit(hours, scores, 1))(hours), color='orange', linestyle='--', label='Best fit line')
plt.title('Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

# Calculate Pearson correlation
r = np.corrcoef(hours, scores)[0,1]
print(f'Pearson correlation coefficient: {r:.2f}')
````
