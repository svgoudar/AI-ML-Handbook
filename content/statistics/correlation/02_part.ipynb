## Correlation

Correlation is a statistical measure that expresses the strength and direction of a relationship between two quantitative variables.

---

### Key Points

1. **Direction**

   * **Positive correlation** → Both variables increase or decrease together.
   * **Negative correlation** → One variable increases while the other decreases.
   * **Zero correlation** → No consistent relationship.

2. **Strength**

   * Measured using a **correlation coefficient** (commonly Pearson’s $r$).
   * Range: $-1 \leq r \leq +1$.

     * $r = +1$: Perfect positive linear relation.
     * $r = -1$: Perfect negative linear relation.
     * $r = 0$: No linear relation.

3. **Types of Correlation Measures**

   * **Pearson correlation**: Linear relationships, continuous data.
   * **Spearman rank correlation**: Monotonic relationships, ranked/ordinal data.
   * **Kendall’s tau**: Rank-based, robust for small datasets.

4. **Interpretation Example**

   * Correlation between study hours and exam scores is usually **positive**.
   * Correlation between exercise time and body fat percentage is typically **negative**.
   * Correlation between shoe size and IQ is **zero**.

---

### Limitations

* Correlation does **not imply causation**. Two variables may correlate due to coincidence or a third factor.
* Outliers can distort correlation values.
* Pearson’s correlation only captures **linear** relationships, not curves or complex patterns.



````python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.randn(50)
y_pos = x + np.random.randn(50) * 0.5   # positive correlation
y_neg = -x + np.random.randn(50) * 0.5  # negative correlation
y_none = np.random.randn(50)            # no correlation

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Positive correlation
axs[0].scatter(x, y_pos, color='blue')
axs[0].set_title("Positive Correlation")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")

# Negative correlation
axs[1].scatter(x, y_neg, color='red')
axs[1].set_title("Negative Correlation")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")

# No correlation
axs[2].scatter(x, y_none, color='green')
axs[2].set_title("No Correlation")
axs[2].set_xlabel("X")
axs[2].set_ylabel("Y")

plt.tight_layout()
plt.show()
````
