## Kendall’s Tau

**Kendall’s Tau (τ)** is a **non-parametric rank correlation coefficient** that measures the strength and direction of the association between two variables. It is based on the idea of **concordant** and **discordant** pairs.

---

### Definition

* Suppose you have two variables $X$ and $Y$ measured on the same $n$ items.
* For every possible pair of items $(i, j)$:

  * The pair is **concordant** if the ranks of both elements agree (i.e., when $x_i > x_j$ and $y_i > y_j$, or $x_i < x_j$ and $y_i < y_j$).
  * The pair is **discordant** if the ranks disagree (i.e., when $x_i > x_j$ but $y_i < y_j$, or vice versa).
  * If ranks are equal (ties), special handling is needed.

---

### Formula

$$
\tau = \frac{(C - D)}{\frac{1}{2} n (n-1)}
$$

Where:

* $C$ = number of concordant pairs
* $D$ = number of discordant pairs
* $n$ = number of observations

The denominator is the total number of possible pairs.

* **Range:** $-1 \leq \tau \leq 1$

  * $+1$: perfect agreement (all pairs concordant).
  * $-1$: perfect disagreement (all pairs discordant).
  * $0$: no association.

---

### When to Use Kendall’s Tau

* Suitable for **ordinal data** or **ranked variables**.
* More robust than Pearson’s correlation for **non-linear** relationships.
* Preferred over Spearman’s rho when sample size is small or when there are many tied ranks.

---

### Comparison with Other Correlations

* **Pearson’s r**: measures **linear correlation** between numeric variables.
* **Spearman’s rho**: rank-based, measures **monotonic** relationships.
* **Kendall’s tau**: also rank-based, but based on pair comparisons, often considered more **statistically rigorous** for ordinal data.



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kendalltau

# Example data: ordinal/ranked variables
# Let's use the same example as in Spearman's:
students = ['A', 'B', 'C', 'D', 'E']
math = np.array([90, 70, 80, 60, 50])
science = np.array([85, 65, 75, 60, 55])

# Scatter plot of original scores
plt.figure(figsize=(6,4))
plt.scatter(math, science, color='brown', s=80)
for i, name in enumerate(students):
    plt.text(math[i]+0.5, science[i], name)
plt.title('Math Score vs Science Score')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Calculate Kendall's tau correlation
tau, _ = kendalltau(math, science)
print(f"Kendall's tau correlation coefficient: {tau:.2f}")
````
