### Cumulative Distribution Function (CDF)

The **Cumulative Distribution Function (CDF)** is a key concept in probability theory that represents the cumulative probability that a random variable takes a value less than or equal to a given threshold.

**Key Characteristics:**

* **Applicability**:
  The CDF applies to **both discrete and continuous random variables**, making it one of the most versatile tools in probability.

* **Purpose**:
  It expresses how probabilities accumulate across the possible values of a random variable.

* **For Discrete Random Variables**:

  * The CDF at a given point is the **sum of probabilities of all outcomes up to and including that point**.
  * **Example**: Rolling a fair die with outcomes {1, 2, 3, 4, 5, 6}.

    * $F(3) = Pr(X \leq 3) = Pr(1) + Pr(2) + Pr(3) = 3 \times \tfrac{1}{6} = 0.5$.
    * $F(6) = Pr(X \leq 6) = 1$, representing the certainty that the die roll will be one of its possible outcomes.

* **For Continuous Random Variables**:

  * The CDF gives the **probability of the random variable falling within the range from the minimum up to a given value**.
  * **Example**: If $Pr(X \leq 40) = 0.8$, it means there is an 80% chance that $X$ is 40 or less. Similarly, $Pr(X \leq 45) = 0.95$ means there is a 95% chance that $X$ is 45 or less.

* **Relationship with PDF**:

  * For continuous distributions, the **PDF (Probability Density Function)** is the **derivative of the CDF**:

    $$
    f(x) = \frac{d}{dx}F(x)
    $$
  * Conversely, the CDF is the integral of the PDF.



**Statistical Measures**

The **CDF** provides the cumulative probability that a random variable takes a value **less than or equal to a specific value**. It is valid for **both discrete and continuous random variables**.


**Key Points of CDF**

1. **Definition**

   * For a random variable $X$, the CDF at a point $x$ is:

   $$
   F(x) = P(X \leq x)
   $$

   This means the CDF accumulates probabilities up to $x$.

2. **Discrete Case**

   * For a dice roll:

   $$
   F(3) = P(X \leq 3) = P(1) + P(2) + P(3)
   $$

3. **Continuous Case**

   * For continuous variables, the CDF is the **integral of the PDF**:

   $$
   F(x) = \int_{-\infty}^x f(t) \, dt
   $$


**Statistical Measures Using CDF**

1. **Mean (Expected Value)**
   The mean can be computed using the **CDF**:

   $$
   E[X] = \int_{0}^{\infty} (1 - F(x)) \, dx - \int_{-\infty}^{0} F(x) \, dx
   $$

   This links the average value of the distribution directly to the CDF.

2. **Median**
   The median is the value $m$ such that:

   $$
   F(m) = 0.5
   $$

   i.e., half of the probability mass lies below $m$.

3. **Mode**
   While the mode is usually derived from the **PDF (continuous)** or **PMF (discrete)**, the **steepest slope in the CDF** indicates where the mode lies.

4. **Variance**
   Variance can be derived using the CDF indirectly:

   $$
   Var(X) = E[X^2] - (E[X])^2
   $$

   and $E[X^2]$ can be computed via:

   $$
   E[X^2] = \int_{0}^{\infty} 2x(1 - F(x)) \, dx
   $$

5. **Percentiles / Quantiles**
   The $p$-th percentile is the value $x_p$ such that:

   $$
   F(x_p) = p
   $$

   Example: The 25th percentile (Q1) satisfies $F(x_{0.25}) = 0.25$.

âœ… **In short:**

* **CDF gives cumulative probability**.
* **Mean, median, variance, and percentiles** can be derived from the CDF.
* The **steepest jump in CDF** indicates the mode.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# --- Discrete Case: Rolling a Fair Die --- #
die_outcomes = np.arange(1, 7)
pmf = np.ones(6) / 6  # Probability of each outcome = 1/6
cdf_discrete = np.cumsum(pmf)

# --- Continuous Case: Normal Distribution --- #
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)  # Standard Normal Distribution PDF
cdf_continuous = norm.cdf(x, loc=0, scale=1)

# --- Plotting --- #
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Discrete Case (Die Roll)
axs[0].bar(die_outcomes, pmf, alpha=0.6, label="PMF (Probabilities)")
axs[0].step(die_outcomes, cdf_discrete, where="mid", color="red", label="CDF (Cumulative)")
axs[0].set_title("Discrete Case: Fair Die Roll", fontsize=14)
axs[0].set_xlabel("Die Outcomes")
axs[0].set_ylabel("Probability / Cumulative Probability")
axs[0].legend()

# Continuous Case (Normal Distribution)
sns.lineplot(x=x, y=pdf, ax=axs[1], label="PDF", color="blue")
sns.lineplot(x=x, y=cdf_continuous, ax=axs[1], label="CDF", color="red")
axs[1].set_title("Continuous Case: Normal Distribution", fontsize=14)
axs[1].set_xlabel("x")
axs[1].set_ylabel("Density / Cumulative Probability")
axs[1].legend()

plt.tight_layout()
plt.show()
```
