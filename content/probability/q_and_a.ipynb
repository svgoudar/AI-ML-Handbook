**Q13. If 5% of population has a disease, test detects correctly 98% of sick, false positive rate 2%. If person tests positive, what is probability he has disease? (Bayes)**

<details><summary>Click to see solution</summary>  

* Let D = disease, T+ = test positive.
* $P(D)=0.05$, $P(\sim D)=0.95$.
* $P(T+|D)=0.98$, $P(T+|\sim D)=0.02$.
* By **Bayesâ€™ Theorem**:

$$
P(D|T+) = \frac{P(T+|D)P(D)}{P(T+|D)P(D)+P(T+|\sim D)P(\sim D)}
$$

Substitute:

$$
= \frac{0.98 \times 0.05}{(0.98 \times 0.05)+(0.02 \times 0.95)} = 0.7206
$$

âœ… **Answer = 72.1%**

</details>
