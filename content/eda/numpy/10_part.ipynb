# Random Sampling in NumPy

NumPy provides a suite of random sampling functions for generating random numbers, sampling from distributions, and shuffling data. This notebook demonstrates the most commonly used random sampling methods in NumPy.

## Key APIs and Concepts in NumPy Random Sampling

- `numpy.random.rand`: Generate samples from a uniform distribution over [0, 1).
- `numpy.random.randn`: Generate samples from the standard normal distribution.
- `numpy.random.randint`: Generate random integers from a specified range.
- `numpy.random.choice`: Randomly select elements from an array.
- `numpy.random.shuffle`: Shuffle the contents of an array in place.
- `numpy.random.permutation`: Return a shuffled copy of an array.
- `numpy.random.seed`: Set the random seed for reproducibility.
- `numpy.random.uniform`: Draw samples from a uniform distribution.
- `numpy.random.normal`: Draw samples from a normal (Gaussian) distribution.

These functions are essential for simulations, bootstrapping, and randomized algorithms in data science and machine learning.

## Visualizing Random Samples

Let's visualize some random samples using matplotlib to better understand their distributions.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
np.random.seed(42)

n = 1000  # Number of samples

# Create 5x3 subplot grid (15 axes, enough for all plots)
fig, axes = plt.subplots(5, 3, figsize=(18, 20))
axes = axes.flatten()  # flatten to 1D for easy indexing

# 1. Uniform Distribution
sns.histplot(np.random.uniform(0, 10, n), kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Uniform Distribution')

# 2. Normal (Gaussian)
sns.histplot(np.random.normal(0, 1, n), kde=True, color='salmon', ax=axes[1])
axes[1].set_title('Normal Distribution (mean=0, std=1)')

# 3. Binomial
sns.histplot(np.random.binomial(10, 0.5, n), bins=range(0, 11), color='green', ax=axes[2])
axes[2].set_title('Binomial (n=10, p=0.5)')

# 4. Poisson
sns.histplot(np.random.poisson(3, n), color='purple', ax=axes[3])
axes[3].set_title('Poisson (λ=3)')

# 5. Exponential
sns.histplot(np.random.exponential(1, n), kde=True, color='orange', ax=axes[4])
axes[4].set_title('Exponential (scale=1)')

# 6. Beta
sns.histplot(np.random.beta(2, 5, n), kde=True, color='brown', ax=axes[5])
axes[5].set_title('Beta (a=2, b=5)')

# 7. Gamma
sns.histplot(np.random.gamma(2, 2, n), kde=True, color='teal', ax=axes[6])
axes[6].set_title('Gamma (shape=2, scale=2)')

# 8. Chi-Square
sns.histplot(np.random.chisquare(3, n), kde=True, color='navy', ax=axes[7])
axes[7].set_title('Chi-Square (df=3)')

# 9. Logistic
sns.histplot(np.random.logistic(0, 1, n), kde=True, color='pink', ax=axes[8])
axes[8].set_title('Logistic Distribution')

# 10. Multinomial
multi = np.random.multinomial(10, [0.2, 0.5, 0.3], n)
sns.histplot(multi[:, 1], color='gray', ax=axes[9])
axes[9].set_title('Multinomial (counts for category 2)')

# 11. Laplace
sns.histplot(np.random.laplace(0, 1, n), kde=True, color='gold', ax=axes[10])
axes[10].set_title('Laplace Distribution')

# 12. Random Integers
sns.histplot(np.random.randint(0, 10, n), bins=range(0, 11), color='lightgreen', ax=axes[11])
axes[11].set_title('Random Integers [0,10)')

# 13. Standard Normal
sns.histplot(np.random.standard_normal(n), kde=True, color='coral', ax=axes[12])
axes[12].set_title('Standard Normal')

# 14. Standard Exponential
sns.histplot(np.random.standard_exponential(n), kde=True, color='indigo', ax=axes[13])
axes[13].set_title('Standard Exponential')

# 15. Standard Gamma
sns.histplot(np.random.standard_gamma(2, n), kde=True, color='cyan', ax=axes[14])
axes[14].set_title('Standard Gamma (shape=2)')

# Adjust layout
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10000 samples from standard normal distribution
data = np.random.randn(10000)

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(data,bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Plot the theoretical normal distribution curve
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'r', linewidth=2)

plt.title("Histogram of randn samples (approx Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Observed vs expected
categories = ['1', '2', '3', '4', '5', '6']
observed = np.array([8, 12, 9, 11, 10, 10])
expected = np.array([10, 10, 10, 10, 10, 10])

x = np.arange(len(categories))  # positions

plt.figure(figsize=(8,5))
plt.bar(x - 0.2, observed, width=0.4, label='Observed', color='skyblue')
plt.bar(x + 0.2, expected, width=0.4, label='Expected', color='orange')

plt.xlabel('Dice Side')
plt.ylabel('Frequency')
plt.title('Chi-Square Goodness of Fit: Dice Roll')
plt.xticks(x, categories)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

### Shuffling Data

You can shuffle arrays or lists in place using NumPy's shuffle function.

````python
from scipy.sparse import csr_matrix

a = csr_matrix([
    [0, 0, 3],
    [0, 4, 0],
    [0, 0, 0]
])

b = csr_matrix([
    [5, 0, 0],
    [0, 0, 0],
    [0, 0, 6]
])

# Sparse addition
c = a + b
print(c)

# Sparse matrix multiplication
d = a.dot(b.transpose())
print(d.toarray())
````
