## Inverse Functions & Identity Functions


* **Definition**: An inverse of a function **reverses** the effect of the original function.

  * If $f : X \to Y$, then $f^{-1} : Y \to X$.
* In simple terms:

  $$
  f(x) = y \quad \implies \quad f^{-1}(y) = x
  $$

**Condition**:

* A function has an inverse **if and only if** it is **bijective**:

  * **Injective (1-to-1)** ‚Üí different inputs map to different outputs.
  * **Surjective (onto)** ‚Üí every element of the codomain is mapped.

---

### Properties of Inverse Functions

1. **Composition Rule**:

   $$
   f(f^{-1}(y)) = y, \quad f^{-1}(f(x)) = x
   $$
2. **Reversibility**: Applying a function and then its inverse (or vice versa) returns the original value.

---

### Identity Function

* **Definition**: The identity function maps every element to itself.

  $$
  I(x) = x
  $$
* **Matrix Representation** (Identity Matrix):

  $$
  I = 
  \begin{bmatrix}
  1 & 0 \\
  0 & 1
  \end{bmatrix}
  \quad \text{(for 2√ó2 case)}
  $$
* **Properties**:

  * **Preservation**: $I(x) = x$.
  * **Linearity**: $I(u+v) = I(u) + I(v)$, and $I(cu) = cI(u)$.
  * **Self-Inverse**: $I^{-1} = I$.

---

### Example: Finding Inverse of a Linear Function

Let

$$
f(x) = 2x + 3
$$

**Step 1: Replace $f(x)$ with $y$:**

$$
y = 2x + 3
$$

**Step 2: Solve for $x$:**

$$
x = \frac{y - 3}{2}
$$

**Step 3: Define inverse function:**

$$
f^{-1}(y) = \frac{y - 3}{2}
$$

‚úÖ Verification:

* $f(f^{-1}(y)) = 2\frac{y-3}{2} + 3 = y$
* $f^{-1}(f(x)) = \frac{(2x+3) - 3}{2} = x$

Thus, $f^{-1}$ is the valid inverse of $f$.

### Applications of Inverse Functions in Data Science



* **Inverse Function**: A transformation that *undoes* another transformation.
* In data science, transformations are frequently applied to data (scaling, encoding, encryption, etc.), and **inverse functions help bring results back to their original interpretable form**.

---

#### Application 1: Normalization & Standardization

**Problem**

Features in a dataset (e.g., house prices) often have **different units and scales**:

* Carpet area (sq. ft.) ‚Üí large numbers (1500, 2000, etc.)
* Number of rooms ‚Üí small integers (1, 2, 3, 4, ‚Ä¶)
* Price ‚Üí lakhs or millions (e.g., 50,00,000)

This difference in magnitude **slows down optimization** and can bias learning.

**Solution:**

We use **feature scaling transformations**:

1. **Standardization (Z-score scaling)**

   $$
   z = \frac{x - \mu}{\sigma}
   $$

   * Converts feature into **zero mean and unit variance** distribution.
   * Helps ML algorithms converge faster.

   **Inverse transformation:**

   $$
   x = z\sigma + \mu
   $$

   (used to recover original values after predictions).

---

2. **Normalization (Min-Max Scaling)**

   $$
   z = \frac{x - \min(x)}{\max(x) - \min(x)}
   $$

   * Rescales data into a fixed range (e.g., $[0,1]$).
   * Useful in **image processing, neural networks**.

   **Inverse transformation:**

   $$
   x = z \cdot (\max(x) - \min(x)) + \min(x)
   $$

---

**Use Case**:

* In **house price prediction**, inputs (carpet area, rooms, land area) are scaled before training.
* Predictions from the trained model (which are in scaled form) are brought back to the **original rupee scale** using the **inverse transformation**.

---

#### Application 2: Distribution Transformations

Many ML models assume **normally distributed data**.
When data is skewed, transformations are applied:

* **Log Transformation**:

  $$
  y = \log(x)
  $$

  * Converts skewed (right-skewed) data into a normal-like distribution.
  * Common in **financial data (income, sales, etc.)**.

  **Inverse transformation:**

  $$
  x = e^y
  $$

* **Box-Cox Transformation** (advanced) ‚Üí also has a defined inverse.

---

#### Application 3: Data Encryption & Decryption

Inverse functions also play a role in **data security**:

* **Encryption function $E(p)$**: converts plain text $p$ into cipher text $c$.
* **Decryption function $D(c)$**: inverse of encryption ‚Üí retrieves $p$.

$$
c = E(p), \quad p = D(c)
$$

**Use Case**: Protecting sensitive ML data such as **medical records, financial transactions**.

---

1. **Summary of Why Inverse Functions Matter in DS**

* üîπ **Scaling & Rescaling**: Standardization and normalization for faster, fairer ML training.
* üîπ **Interpretability**: Predictions are transformed back into **original units** for real-world meaning.
* üîπ **Distribution Fixing**: Log & Box-Cox transformations normalize skewed data.
* üîπ **Security**: Encryption/Decryption relies on inverses.

---

üëâ So, **inverse functions bridge the gap** between raw real-world data and ML-friendly transformations, making models both efficient and interpretable.


### Inverse of a Matrix and Determinant

**Key Concepts**

* A **determinant** is a scalar value that can be computed from a square matrix.
* It gives important information such as:

  * Whether the matrix is invertible (det ‚â† 0).
  * Geometric interpretation: the scaling factor of the linear transformation.
* **Inverse Matrix**:

  * If $A$ is invertible, there exists $A^{-1}$ such that

    $$
    A \cdot A^{-1} = I
    $$
  * Used to ‚Äúreverse‚Äù transformations: $y = Ax$ ‚Üí $x = A^{-1}y$.

---

**Formula for 2√ó2 Matrix**

For

$$
A = \begin{bmatrix} a & b \\ c & d \end{bmatrix},
$$

* **Determinant**:

$$
\det(A) = ad - bc
$$

* **Inverse**:

$$
A^{-1} = \frac{1}{\det(A)} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$

Condition: $\det(A) \neq 0$.

---

  **Example Problem**

Given

$$
A = \begin{bmatrix} 4 & 7 \\ 2 & 6 \end{bmatrix}
$$

#### Step 1: Determinant

$$
\det(A) = (4)(6) - (7)(2) = 24 - 14 = 10
$$

Since det ‚â† 0, $A$ is invertible.

---

#### Step 2: Inverse

$$
A^{-1} = \frac{1}{10} 
\begin{bmatrix} 6 & -7 \\ -2 & 4 \end{bmatrix}
=
\begin{bmatrix} 0.6 & -0.7 \\ -0.2 & 0.4 \end{bmatrix}
$$

---

#### Step 3: Verification with a Vector

Let

$$
x = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

* Apply transformation:

$$
y = A \cdot x =
\begin{bmatrix} 4 & 7 \\ 2 & 6 \end{bmatrix}
\begin{bmatrix} 1 \\ 1 \end{bmatrix}
=
\begin{bmatrix} 11 \\ 8 \end{bmatrix}
$$

* Recover original vector:

$$
x = A^{-1} y =
\begin{bmatrix} 0.6 & -0.7 \\ -0.2 & 0.4 \end{bmatrix}
\begin{bmatrix} 11 \\ 8 \end{bmatrix}
=
\begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

‚úÖ Verified: $A^{-1}$ correctly recovers the original vector.

---

**Conclusion**

* Determinant tells us if a matrix is invertible.
* Inverse matrices undo linear transformations.
* Verified using a simple vector example.


````python
import matplotlib.pyplot as plt
import numpy as np

# Define matrix A and its inverse
A = np.array([[4, 7],
              [2, 6]])
A_inv = np.linalg.inv(A)

# Define original vector
x = np.array([1, 1])

# Apply transformation
y = A @ x

# Apply inverse transformation
x_recovered = A_inv @ y

# Plot vectors
plt.figure(figsize=(7,7))
ax = plt.gca()
ax.set_xlim(-1, 12)
ax.set_ylim(-1, 12)

# Plot original vector
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original x')

# Plot transformed vector
plt.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed y = A¬∑x')

# Plot recovered vector
plt.quiver(0, 0, x_recovered[0], x_recovered[1], angles='xy', scale_units='xy', scale=1, color='green', label='Recovered x = A‚Åª¬π¬∑y')

# Add labels and grid
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Inverse Transformation Visualization")

plt.show()
````
