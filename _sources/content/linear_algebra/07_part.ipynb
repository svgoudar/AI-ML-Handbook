## Eigenvectors and Eigenvalues


### Linear Transformation

* Transformation of unit vectors $i$ and $j$ using a matrix.
* Observed how the vectors move to new coordinates after transformation.
* Demonstrated the **determinant** as the factor by which the area (or volume) changes.

---

### Eigenvectors

* Defined as vectors whose **direction does not change** under a linear transformation.
* Can be **scaled** (stretched or compressed) but not rotated.
* Visualization: vectors that, after transformation, lie on the same line as the original.

---

### Eigenvalues

* Scalars that indicate **how much an eigenvector is stretched or compressed**.
* Example: If a vector scales from length 1 to 6 after transformation, the eigenvalue is 6.

---

### Identifying Eigenvectors

* Normal vectors change direction after transformation.
* Eigenvectors retain direction but may change magnitude.
* Multiple examples demonstrated visually.

---

### Mathematical Formulation

* **Eigenvalue equation:** $A \mathbf{v} = \lambda \mathbf{v}$

  * $A$ = transformation matrix
  * $\mathbf{v}$ = eigenvector
  * $\lambda$ = eigenvalue

* **Finding eigenvalues:**

  1. Solve $\det(A - \lambda I) = 0$
  2. Use quadratic formula for 2×2 matrices.

* **Finding eigenvectors:**

  1. Solve $(A - \lambda I) \mathbf{v} = 0$
  2. Determine vector components from the resulting equations.

* Example: For matrix $A = \begin{bmatrix}4 & 2 \\ 1 & 3\end{bmatrix}$

  * Eigenvalues: $\lambda_1 = 5$, $\lambda_2 = 2$
  * Corresponding eigenvectors:

    * $v_1 = [1, 1]$
    * $v_2 = [1, -2]$

---

### Interpretation

* Eigenvectors indicate **directions** of scaling.
* Eigenvalues indicate **factor of scaling** along those directions.
* Fundamental in applications like **Principal Component Analysis (PCA)** for **dimensionality reduction**.

---

**8. Key Takeaways**

* Eigenvectors ≠ normal vectors; they retain direction under transformation.
* Eigenvalues quantify the magnitude of stretching/compression.
* Practical applications in **data science**, especially in **PCA**.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define matrix A (transformation)
A = np.array([[4, 2],
              [1, 3]])

# Define unit vectors i and j
i = np.array([1, 0])
j = np.array([0, 1])

# Transform unit vectors
i_trans = A @ i
j_trans = A @ j

# Define eigenvectors manually for visualization
eig_vec1 = np.array([1, 1])
eig_vec2 = np.array([1, -2])
eig_val1 = 5
eig_val2 = 2

# Scale eigenvectors for visualization
eig_vec1_scaled = eig_val1 * eig_vec1
eig_vec2_scaled = eig_val2 * eig_vec2

# Plot original vectors
plt.quiver(0, 0, i[0], i[1], angles='xy', scale_units='xy', scale=1, color='blue', label='i (unit vector)')
plt.quiver(0, 0, j[0], j[1], angles='xy', scale_units='xy', scale=1, color='green', label='j (unit vector)')

# Plot transformed vectors
plt.quiver(0, 0, i_trans[0], i_trans[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='i transformed')
plt.quiver(0, 0, j_trans[0], j_trans[1], angles='xy', scale_units='xy', scale=1, color='lime', label='j transformed')

# Plot eigenvectors and their scaling
plt.quiver(0, 0, eig_vec1_scaled[0], eig_vec1_scaled[1], angles='xy', scale_units='xy', scale=1, color='red', label='Eigenvector 1 (λ=5)')
plt.quiver(0, 0, eig_vec2_scaled[0], eig_vec2_scaled[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Eigenvector 2 (λ=2)')

# Set plot limits and grid
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Linear Transformation, Eigenvectors, and Eigenvalues')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the transformation matrix
A = np.array([[4, 2],
              [1, 3]])

# Original grid of points for visualization
x = np.array([0, 1, 0, 1])
y = np.array([0, 0, 1, 1])
points = np.vstack([x, y])

# Apply linear transformation
transformed_points = A @ points

# Plot original square
plt.figure(figsize=(8, 8))
plt.plot(points[0, [0,1,3,2,0]], points[1, [0,1,3,2,0]], 'b-', label='Original Square')
plt.scatter(points[0, :], points[1, :], color='blue')

# Plot transformed parallelogram
plt.plot(transformed_points[0, [0,1,3,2,0]], transformed_points[1, [0,1,3,2,0]], 'r-', label='Transformed Shape')
plt.scatter(transformed_points[0, :], transformed_points[1, :], color='red')

# Draw eigenvectors
eigvals, eigvecs = np.linalg.eig(A)
for i in range(len(eigvals)):
    vec = eigvecs[:, i]
    plt.quiver(0, 0, vec[0]*eigvals[i], vec[1]*eigvals[i], angles='xy', scale_units='xy', scale=1, color='green', label=f'Eigenvector {i+1}' if i==0 else None)

plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Linear Transformation with Eigenvectors')
plt.legend()
plt.grid(True)
plt.show()
````
