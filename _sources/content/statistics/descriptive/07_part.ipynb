# Histogram and Skewness

## Histogram

* **Definition**:
  A histogram is a graphical representation of the distribution of numerical data.

  * Shows **shape, central tendency, and variability**.
  * Approximates the **probability distribution** of a continuous variable.

* **Steps to Construct a Histogram**:

  1. Identify the range of data (min → max).
  2. Decide number of **bins** (e.g., 10 bins).
  3. Compute **bin size** = (max – min) ÷ bins.
  4. Count frequency of values in each bin.
  5. Plot bins on the x-axis and frequency on the y-axis.

* **Relation to Probability Distribution**:

  * Histogram shows **empirical distribution**.
  * By smoothing (using **Kernel Density Estimation, KDE**), we get the **Probability Density Function (PDF)**.

---

##  Skewness

Skewness shows how data is distributed around the mean.

### Symmetrical Distribution (Normal/Gaussian)

* Histogram forms a **bell curve**.
* Left and right sides are mirror images.
* **Mean = Median = Mode**.
* In Boxplot: median lies exactly in the center.
* No skewness.

---

### Right-Skewed (Positive Skew)

* Tail extends to the **right**.
* Outliers are on the **higher values** side.
* Example: **Log-normal distribution**.
* Relationship: **Mean > Median > Mode**.
* Boxplot: median shifts left, Q3–Q2 > Q2–Q1.

---

### Left-Skewed (Negative Skew)

* Tail extends to the **left**.
* Outliers are on the **lower values** side.
* Relationship: **Mean < Median < Mode**.
* Boxplot: median shifts right, Q2–Q1 > Q3–Q2.

---

### Key Takeaways

* **Histogram** shows frequency distribution; **KDE smoothing** gives PDF.
* **Symmetrical distribution** → Mean = Median = Mode.
* **Right Skew** → Mean > Median > Mode.
* **Left Skew** → Mean < Median < Mode.
* **Boxplots** help visualize skewness and detect outliers.



````python
import matplotlib.pyplot as plt
import numpy as np

# Function to generate distributions and plots
def plot_skewness_examples():
    np.random.seed(42)
    
    # Generate distributions
    normal = np.random.normal(loc=0, scale=1, size=1000)
    right_skew = np.random.exponential(scale=1, size=1000)
    left_skew = -np.random.exponential(scale=1, size=1000)
    
    data = [normal, right_skew, left_skew]
    titles = ["Symmetrical (Normal)", "Right-Skewed (Positive)", "Left-Skewed (Negative)"]
    
    fig, axes = plt.subplots(3, 2, figsize=(12, 10))
    
    for i in range(3):
        # Histogram
        axes[i, 0].hist(data[i], bins=30, edgecolor="black", alpha=0.7, density=True)
        axes[i, 0].set_title(f"Histogram: {titles[i]}")
        axes[i, 0].axvline(np.mean(data[i]), color='red', linestyle='--', label="Mean")
        axes[i, 0].axvline(np.median(data[i]), color='green', linestyle='-', label="Median")
        axes[i, 0].legend()
        
        # Boxplot
        axes[i, 1].boxplot(data[i], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
        axes[i, 1].set_title(f"Boxplot: {titles[i]}")
    
    plt.tight_layout()
    plt.show()

plot_skewness_examples()
````
