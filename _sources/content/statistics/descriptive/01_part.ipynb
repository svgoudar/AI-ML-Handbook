## Measures of Central Tendency

**Definition**

* Statistical metrics that describe the **center point** of a dataset.
* Provide a single value representing the central position (e.g., ages → mean age).

---

### Types

1. **Mean (Average)**

   * Formula:

     * Population Mean: \$\mu = \frac{\sum\_{i=1}^{N} x\_i}{N}\$
     * Sample Mean: \$\bar{x} = \frac{\sum\_{i=1}^{n} x\_i}{n}\$
   * **Characteristics**:

     * Affected by extreme **outliers**.
     * Used for **interval & ratio data**.
   * Example: Numbers (1,2,3,4,5) → mean = 3.
     Adding outlier (100) → mean jumps to 23.

---

2. **Median (Middle Value)**

   * Definition: Middle value when data is arranged in ascending/descending order.
   * Odd count → single middle element.
   * Even count → average of two middle elements.
   * **Characteristics**:

     * **Not affected** by outliers.
     * Used for **ordinal, interval, ratio data**.
   * Example: (1,2,3,4,5) → median = 3.
     Adding 100 → median only shifts slightly (3.5).

---

3. **Mode (Most Frequent Value)**

   * Definition: Value occurring **most frequently** in the dataset.
   * Types:

     * Unimodal (one mode).
     * Bimodal (two modes).
     * Multimodal (multiple modes).
   * **Characteristics**:

     * Not affected by outliers.
     * Used for **all data types** (nominal, ordinal, interval, ratio).
   * Example: (6,7,7,7,9) → mode = 7.

---

### When to Use

* **Mean** → Best for **symmetrically distributed data without outliers**.
* **Median** → Best for **skewed data / with outliers**.
* **Mode** → Best for **categorical (nominal) data**.

---

### Real-World Application: Feature Engineering

* **Handling missing values**:

  * If data is symmetric → replace with **mean**.
  * If data is skewed/outliers → replace with **median**.
  * If data is categorical (e.g., gender, degree) → replace with **mode**.

---

**Key Takeaway**:

* **Mean, Median, Mode** each have strengths depending on data type and presence of outliers.
* Widely used in **data preprocessing, feature engineering, and statistical analysis**.


````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats

# Generate datasets
np.random.seed(42)

# Symmetric (Normal Distribution)
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Right-Skewed (Gamma Distribution)
data_right = np.random.gamma(shape=2, scale=2, size=1000)

# Left-Skewed (Take negative of a right-skewed distribution)
data_left = -np.random.gamma(shape=2, scale=2, size=1000)

# Function to calculate and plot Mean, Median, Mode
def plot_distribution(ax, data, title):
    mean_val = np.mean(data)
    median_val = np.median(data)
    mode_result = stats.mode(np.round(data, 1), keepdims=True)
    mode_val = mode_result.mode[0]

    sns.kdeplot(data, fill=True, color='skyblue', alpha=0.5, ax=ax)
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.2f}')
    ax.axvline(median_val, color='green', linestyle='-', linewidth=2, label=f'Median = {median_val:.2f}')
    ax.axvline(mode_val, color='orange', linestyle='-', linewidth=2, label=f'Mode ≈ {mode_val:.2f}')
    
    ax.set_title(title)
    ax.legend()

# Plot all three distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

plot_distribution(axes[0], data_normal, "Symmetric Distribution (Normal)")
plot_distribution(axes[1], data_right, "Right-Skewed Distribution (Gamma)")
plot_distribution(axes[2], data_left, "Left-Skewed Distribution (Negative Gamma)")

plt.tight_layout()
plt.show()
````
