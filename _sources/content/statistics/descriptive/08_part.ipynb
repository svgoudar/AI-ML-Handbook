## Covariance

* **Definition**: Measures how much two random variables change together.

  * Positive covariance → when X increases, Y also increases (and vice versa).
  * Negative covariance → when X increases, Y decreases.

* **Formula**:

  $$
  \text{Cov}(X, Y) = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{n-1}
  $$

  * If $X = Y$, covariance reduces to **variance**.

* **Example**:

  * Hours studied vs. Exam scores → Positive covariance.
  * If X ↑ and Y ↓ → Negative covariance.

* **Advantages**: Quantifies the relationship between two variables.

* **Disadvantages**: Value is **unbounded** (−∞ to +∞), making comparisons difficult.

---

## Correlation

* Solves covariance’s limitation by **scaling values** between **−1 to +1**.
* **Formula (Pearson correlation coefficient)**:

  $$
  r = \frac{\text{Cov}(X,Y)}{\sigma_X \cdot \sigma_Y}
  $$

  * +1 → Perfect positive correlation.
  * −1 → Perfect negative correlation.
  * 0 → No linear relationship.

### Pearson Correlation

* Measures **linear relationship** between X and Y.
* Works well when data points form a straight-line relationship.
* Limitation: Cannot capture **non-linear relationships**.

### Spearman Rank Correlation

* Uses **ranks of data** instead of raw values.
* Formula:

  $$
  r_s = \frac{\text{Cov}(\text{rank}(X), \text{rank}(Y))}{\sigma_{\text{rank}(X)} \cdot \sigma_{\text{rank}(Y)}}
  $$
* Captures **monotonic relationships** (both linear and non-linear).
* Example: Even if Y increases at a slower/faster rate than X, Spearman still shows strong correlation.

---

### Use Cases in Data Science

* **Feature selection**:

  * Strongly correlated features with target → keep.
  * Features with correlation ≈ 0 → remove (not useful).
  * Positive correlation (e.g., house size ↔ price).
  * Negative correlation (e.g., haunted house ↔ price).
* **EDA**: Understanding variable relationships before modeling.

---

✅ **Key Takeaways**

* Covariance: Direction of relationship (+/−).
* Correlation: Direction + Strength (scaled to −1 to +1).
* Pearson: Linear only.
* Spearman: Linear + Non-linear monotonic.



````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Sample data for visualization
np.random.seed(42)

# Positive covariance (X and Y grow together)
x_pos = np.linspace(1, 10, 50)
y_pos = x_pos + np.random.normal(0, 1, 50)

# Negative covariance (X increases, Y decreases)
x_neg = np.linspace(1, 10, 50)
y_neg = -x_neg + np.random.normal(0, 1, 50)

# No correlation (independent random values)
x_none = np.linspace(1, 10, 50)
y_none = np.random.uniform(-10, 10, 50)

# Spearman monotonic non-linear relationship
x_spear = np.linspace(1, 10, 50)
y_spear = np.log(x_spear) + np.random.normal(0, 0.1, 50)

# Plot
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Positive Covariance/Correlation
sns.scatterplot(x=x_pos, y=y_pos, ax=axs[0, 0], color="green")
axs[0, 0].set_title("Positive Covariance/Correlation")
axs[0, 0].set_xlabel("X")
axs[0, 0].set_ylabel("Y")

# Negative Covariance/Correlation
sns.scatterplot(x=x_neg, y=y_neg, ax=axs[0, 1], color="red")
axs[0, 1].set_title("Negative Covariance/Correlation")
axs[0, 1].set_xlabel("X")
axs[0, 1].set_ylabel("Y")

# No Correlation
sns.scatterplot(x=x_none, y=y_none, ax=axs[1, 0], color="blue")
axs[1, 0].set_title("No Correlation")
axs[1, 0].set_xlabel("X")
axs[1, 0].set_ylabel("Y")

# Spearman Monotonic Non-linear
sns.scatterplot(x=x_spear, y=y_spear, ax=axs[1, 1], color="purple")
axs[1, 1].set_title("Spearman (Monotonic Non-linear)")
axs[1, 1].set_xlabel("X")
axs[1, 1].set_ylabel("Y")

plt.tight_layout()
plt.show()
````
