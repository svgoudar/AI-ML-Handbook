# Measures of Central Tendency

Measures of central tendency describe the **“center”** or typical value of a dataset. The main measures are **Mean**, **Median**, and **Mode**.


### Mean (Average)

* **Definition:** The sum of all data points divided by the total number of points.
* **Formula:**

$$
\text{Mean} = \frac{\sum x_i}{n}
$$

* **Properties:**

  1. Uses all observations.
  2. Sensitive to extreme values (outliers).
  3. Unique for a given dataset.

* **Applications:**

  * Economic indicators (e.g., average income)
  * Exam scores analysis
  * Scientific measurements

* **When to use:**

  * Data is **symmetrically distributed**
  * No extreme outliers

* **Example:** Dataset: 5, 7, 3, 8, 10

$$
\text{Mean} = \frac{5+7+3+8+10}{5} = 6.6
$$

---

### Median

* **Definition:** The middle value in a sorted dataset.

* **Steps:**

  1. Arrange data in ascending order.
  2. Odd number of values → middle value
  3. Even number of values → average of two middle values

* **Properties:**

  1. Not affected by extreme values (robust).
  2. Divides data into two equal halves.
  3. Unique for a given dataset.

* **Applications:**

  * Household income analysis (to avoid outlier effects)
  * Real estate prices
  * Skewed distributions

* **When to use:**

  * Data is **skewed** or contains outliers

* **Example:** Dataset: 3, 5, 7, 8, 10 → **Median = 7**
  Dataset: 3, 5, 7, 8 → **Median = (5+7)/2 = 6**

---

### Mode

* **Definition:** The most frequently occurring value in the dataset.

* **Properties:**

  1. Can be **more than one value** (bimodal or multimodal).
  2. Can be used for **categorical data**.
  3. Not affected by extreme values.

* **Applications:**

  * Market research (most popular product)
  * Education (most common grade)
  * Categorical data analysis

* **When to use:**

  * Data is **categorical**
  * Identify most common/repeated values

* **Example:** Dataset: 3, 5, 7, 7, 10 → **Mode = 7**

---

### Python Demonstration

```python
import statistics

data = [5, 7, 3, 8, 10]

mean_value = statistics.mean(data)
median_value = statistics.median(data)
mode_value = statistics.mode([3, 5, 7, 7, 10])

mean_value, median_value, mode_value
```

**Output:**

```
Mean: 6.6, Median: 7, Mode: 7
```

---

✅ Summary of Use Cases:

| Measure | Best for                  | Sensitive to Outliers | Data Type             |
| ------- | ------------------------- | --------------------- | --------------------- |
| Mean    | Symmetric data            | Yes                   | Numeric               |
| Median  | Skewed/outlier-prone data | No                    | Numeric               |
| Mode    | Most frequent value       | No                    | Numeric & Categorical |

If you like, I can also create a **diagram showing mean, median, and mode on a skewed dataset** to visualize their differences and practical use.

Do you want me to do that?


````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats

# Continuous dataset example
data = np.random.gamma(shape=2, scale=2, size=1000)  # Skewed continuous data

# Calculate Mean, Median, Mode
mean_val = np.mean(data)
median_val = np.median(data)
mode_result = stats.mode(np.round(data, 1), keepdims=True)  # rounding to estimate mode for continuous data
mode_val = mode_result.mode[0]

# Plot KDE
plt.figure(figsize=(10,5))
sns.kdeplot(data, fill=True, color='skyblue', alpha=0.5)

# Add lines for Mean, Median, Mode
plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='-', linewidth=2, label=f'Median = {median_val:.2f}')
plt.axvline(mode_val, color='orange', linestyle='-', linewidth=2, label=f'Mode ≈ {mode_val:.2f}')

plt.title('Mean, Median, and Mode on a Continuous Skewed Dataset (KDE)')
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.legend()
plt.show()
````
