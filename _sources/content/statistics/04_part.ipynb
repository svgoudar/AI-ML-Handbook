## Time Series (TS)

A **time series** is a sequence of observations recorded **over time** at **regular intervals** (seconds, minutes, hours, days, months, years). The key idea is that the **order of data points matters**.

**Examples:**

* Daily stock prices of a company.
* Monthly sales of a product.
* Annual rainfall in a city.
* Hourly temperature readings.

### Components of a Time Series

Time series can usually be decomposed into four components:

1. **Trend (T)**

   * The long-term movement or general direction in the data.
   * Example: A steady increase in annual GDP over 10 years.

2. **Seasonality (S)**

   * Repeating short-term patterns, usually within a year or day/week.
   * Example: Ice cream sales peak every summer.

3. **Cyclic (C)**

   * Fluctuations over longer periods, often tied to economic or business cycles.
   * Example: Recessions or booms affecting sales over several years.

4. **Irregular / Random (I)**

   * Noise or unpredictable variations.
   * Example: Sudden storm affecting sales on a particular day.

So mathematically, we can express a series as:

* **Additive model:** $Y_t = T_t + S_t + C_t + I_t$
* **Multiplicative model:** $Y_t = T_t \times S_t \times C_t \times I_t$

Additive is used when components are roughly **constant**, multiplicative when components **grow/shrink proportionally**.

### Statistical Forecasting

**Forecasting** is the process of predicting future values based on **past observations**. In statistics, we aim to model the underlying patterns (trend, seasonality, cycles) and project them forward.

### Steps in Statistical Forecasting

1. **Visualize the Data**

   * Plot the time series to identify trends, seasonality, and anomalies.

2. **Stationarity Check**

   * A stationary series has constant mean, variance, and autocorrelation over time.
   * Many forecasting methods require stationarity.
   * Use **Dickey-Fuller test** or **KPSS test** for checking stationarity.

3. **Transform / Detrend / Deseasonalize**

   * Remove trend or seasonality if required to make series stationary.

4. **Model Selection**

   * Classical statistical models:

     * **AR (Autoregressive)**: $X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} + ... + \epsilon_t$
     * **MA (Moving Average)**: $X_t = \epsilon_t + \theta_1 \epsilon_{t-1} + ...$
     * **ARMA / ARIMA**: Combines AR and MA, with differencing for trend.
     * **Seasonal ARIMA (SARIMA)**: ARIMA + seasonal component.
     * **Exponential Smoothing (ETS / Holt-Winters)**: Weights recent observations more.

5. **Model Fitting & Validation**

   * Split data into training and test sets.
   * Fit model to training set and forecast on test set.
   * Measure accuracy using metrics like:

     * MAE (Mean Absolute Error)
     * RMSE (Root Mean Square Error)
     * MAPE (Mean Absolute Percentage Error)

6. **Forecasting Future Values**

   * Once the model performs well on historical data, predict future points with confidence intervals.

---

### Key Concepts in Time Series Forecasting

* **Lag:** Previous observations used to predict current/future values.
* **Autocorrelation (ACF):** Correlation of series with lagged versions of itself.
* **Partial Autocorrelation (PACF):** Measures correlation of series with its lag after removing effect of shorter lags.
* **Seasonal Decomposition:** Separating trend, seasonal, and residual components.

---

### Example

Suppose monthly sales data for 5 years:

* Plot shows upward trend → use ARIMA with differencing.
* Seasonal spikes in December → include seasonal term (SARIMA).
* Forecast next 12 months → gives predicted sales and confidence intervals.

---

In short:

* **Time Series:** Data observed over time.
* **Forecasting:** Statistical modeling to predict future based on past patterns.
* **Goal:** Identify trend, seasonality, cycles, remove randomness, and predict reliably.



````python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic time series data
np.random.seed(42)
n = 120  # 10 years of monthly data
t = np.arange(n)

# Components
trend = 0.2 * t
seasonality = 10 * np.sin(2 * np.pi * t / 12)
noise = np.random.normal(0, 2, n)
series = 50 + trend + seasonality + noise

plt.figure(figsize=(12, 7))
plt.plot(t, series, label='Observed Series', color='black')
plt.plot(t, 50 + trend, label='Trend', color='red', linestyle='--')
plt.plot(t, 50 + seasonality, label='Seasonality', color='blue', linestyle=':')
plt.title('Synthetic Time Series: Trend + Seasonality + Noise')
plt.xlabel('Month')
plt.ylabel('Value')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````
