## Covariance

Mathematically, covariance is the **expected product of deviations** of two variables from their means:

$$
\text{Cov}(X, Y) = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
$$

* $X_i, Y_i$: data points
* $\bar{X}, \bar{Y}$: means of variables
* $n$: number of observations

---

### Interpretation

* **Positive covariance** → when $X$ is above its mean, $Y$ tends to be above its mean (move in same direction).
* **Negative covariance** → when $X$ is above its mean, $Y$ tends to be below its mean (move in opposite directions).
* **Zero covariance** → no linear relationship.

---

### Relation to Correlation

* Covariance gives the **direction** of the relationship but not the **strength**.
* Correlation is a **standardized covariance**:

$$
r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
$$

where $\sigma_X, \sigma_Y$ are standard deviations.
Thus, correlation is **scale-free**, while covariance depends on units.

---

### Example

* If we measure **height (cm)** and **weight (kg)**:

  * Positive covariance → taller people tend to weigh more.
* If we measure **speed of car (km/h)** and **time to reach destination (min)**:

  * Negative covariance → higher speed leads to lower time.



````python
import numpy as np
import matplotlib.pyplot as plt

# Example data: height (cm) and weight (kg)
np.random.seed(0)
height = np.random.normal(170, 10, 100)
weight = 0.5 * height + np.random.normal(0, 5, 100)

# Scatter plot
plt.figure(figsize=(7,5))
plt.scatter(height, weight, alpha=0.7)
plt.title('Scatter Plot: Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Calculate covariance and correlation
cov = np.cov(height, weight)[0,1]
corr = np.corrcoef(height, weight)[0,1]
print(f'Covariance: {cov:.2f}')
print(f'Correlation: {corr:.2f}')
````
