##  Percentage vs Percentile

* **Percentage** ‚Üí Out of 100, ‚Äúwhat portion‚Äù of data fits a condition.

  * Example: Odd numbers in {1,2,3,4,5,6} ‚Üí $3/6 \times 100 = 50\%$.

* **Percentile** ‚Üí A value below which a certain % of data lies.

  * Example: 90th percentile means 90% of the data is **below** that value.

---

### Percentile Calculation

* **Formula (Percentile of a value x):**

  $$
  \text{Percentile}(x) = \frac{\# \text{values below x}}{n} \times 100
  $$

* Example dataset: {2,2,3,4,5,5,6,7,8,8,8,9,9,10}

  * Percentile of 9 ‚Üí 11 values below 9, total $n=14$.

  $$
  \frac{11}{14}\times100 = 78.57
  $$

  ‚úÖ Meaning: 78.57% of values lie below 9.

---

### Finding the Value for a Given Percentile

* Formula:

  $$
  P = \frac{k}{100} \times (n+1)
  $$

  where $k$ = desired percentile (25, 50, 75 ‚Ä¶), $n$ = total data points.

* If $P$ is not an integer ‚Üí interpolate (average between closest ranks).

* Example: Find **25th percentile** of dataset ($n=14$):

  $$
  P = \frac{25}{100} \times (14+1) = 3.75
  $$

  ‚Üí Between 3rd (value=3) and 4th (value=4).
  ‚Üí Average = 3.5.
  ‚úÖ 25th percentile = **3.5**.

---

### Quartiles

* Quartiles are just **special percentiles** that split data into four parts:

  * **Q1 (First Quartile)** = 25th percentile
  * **Q2 (Second Quartile / Median)** = 50th percentile
  * **Q3 (Third Quartile)** = 75th percentile

üëâ Each quartile represents a **quarter of the data distribution**.

---

‚úÖ So:

* **Percentile ‚Üí position in distribution.**
* **Quartiles ‚Üí specific percentiles (25%, 50%, 75%).**



````python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = [2, 2, 3, 4, 5, 5, 6, 7, 8, 8, 8, 9, 9, 10]

# Compute quartiles
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # Median
Q3 = np.percentile(data, 75)
minimum = np.min(data)
maximum = np.max(data)

# Create boxplot
plt.figure(figsize=(8,6))
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="skyblue", alpha=0.5))

# Add indicators for quartiles and other statistics
plt.axvline(Q1, color='green', linestyle='--', label=f'Q1 (25th) = {Q1:.2f}')
plt.axvline(Q2, color='red', linestyle='-', label=f'Q2 (Median) = {Q2:.2f}')
plt.axvline(Q3, color='orange', linestyle='--', label=f'Q3 (75th) = {Q3:.2f}')
plt.axvline(minimum, color='purple', linestyle=':', label=f'Min = {minimum}')
plt.axvline(maximum, color='brown', linestyle=':', label=f'Max = {maximum}')

# Titles and labels
plt.title("Boxplot with Quartile Indicators")
plt.xlabel("Values")
plt.legend()
plt.show()
````
