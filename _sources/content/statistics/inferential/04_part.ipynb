

## Student’s t-Distribution

1. **When to Use t-Distribution**

   * In **Z-tests**, we require the **population standard deviation (σ)**.
   * Often, **σ is unknown**, which makes Z-tests inapplicable.
   * In such cases, we use **Student’s t-distribution** with the **sample standard deviation (s)**.

2. **t-Test Formula**

   $$
   t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
   $$

   * $\bar{X}$ = sample mean
   * μ = population mean (hypothesized)
   * s = **sample standard deviation**
   * n = sample size

3. **Degree of Freedom (DOF)**

   * Important parameter for t-distribution.
   * Formula:

     $$
     \text{DOF} = n - 1
     $$
   * Example: For 3 people and 3 chairs, only 2 people have options → DOF = 2.

4. **t-Table**

   * Similar to Z-table but depends on **DOF**.
   * Provides **critical values** for **one-tailed** or **two-tailed tests** based on **confidence levels**.
   * Used to determine if we **reject or fail to reject the null hypothesis**.

5. **Key Points**

   * t-distribution is used when **population standard deviation is unknown**.
   * Sample standard deviation replaces population standard deviation.
   * DOF adjusts the shape of the t-distribution.
   * t-table helps in hypothesis testing (one-tailed and two-tailed).



## Problem Statement

A nutritionist claims that a new diet plan helps reduce weight by an average of **5 kg** in one month. To verify this, a sample of **10 participants** follows the diet plan for a month. At the end of the month, the sample shows an **average weight loss of 4.2 kg** with a **sample standard deviation of 0.8 kg**.

**Question:** At a **5% significance level**, can we conclude that the diet plan reduces weight by **less than the claimed 5 kg**?

**Notes:**

* Population standard deviation is **unknown**.
* Use **Student’s t-test** (one-tailed, left-tailed test).

---

## Solution

### Step 1: Set Up Hypotheses

Since we want to test if the **weight loss is less than 5 kg**, this is a **left-tailed t-test**:

$$
H_0: \mu = 5 \quad (\text{Diet plan reduces 5 kg})
$$

$$
H_1: \mu < 5 \quad (\text{Diet plan reduces less than 5 kg})
$$

---

### Step 2: Determine Test Statistic

The **t-test formula**:

$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$

Where:

* $\bar{X} = 4.2$ (sample mean)
* μ = 5 (claimed mean)
* s = 0.8 (sample standard deviation)
* n = 10 (sample size)

---

### Step 3: Calculate t-value

1. Standard error (SE):

$$
SE = \frac{s}{\sqrt{n}} = \frac{0.8}{\sqrt{10}} \approx \frac{0.8}{3.162} \approx 0.253
$$

2. t-value:

$$
t = \frac{4.2 - 5}{0.253} = \frac{-0.8}{0.253} \approx -3.16
$$

---

### Step 4: Determine Critical t-value

* Degrees of Freedom: $df = n - 1 = 10 - 1 = 9$
* Significance level: α = 0.05, left-tailed test
* From **t-table**, critical t-value $t_{critical} \approx -1.833$

---

### Step 5: Compare t-value with t-critical

$$
t_{calculated} = -3.16 \quad \text{vs} \quad t_{critical} = -1.833
$$

* Since **t\_calculated < t\_critical**, we **reject the null hypothesis**.

---

### Step 6: Conclusion

At **5% significance level**, there is **enough evidence** to conclude that the diet plan reduces weight by **less than the claimed 5 kg**.

---

✅ **Summary Table**

| Parameter               | Value     |
| ----------------------- | --------- |
| Sample mean ($\bar{X}$) | 4.2       |
| Claimed mean (μ)        | 5         |
| Sample std dev (s)      | 0.8       |
| Sample size (n)         | 10        |
| Degrees of freedom      | 9         |
| t\_calculated           | -3.16     |
| t\_critical (α=0.05)    | -1.833    |
| Decision                | Reject H0 |




## Problem Statement

A team of researchers wants to test a new medication to see if it has an effect on intelligence. The population average IQ is **100**. A sample of **30 participants** who took the medication has a mean IQ of **140** with a **sample standard deviation of 20**.

**Question:** At a **95% confidence level (α = 0.05)**, can we conclude that the medication affects intelligence?

* **Notes:**

  * Population standard deviation is unknown → use **t-test**
  * Effect could be **positive or negative** → **two-tailed test**

---

## Solution Summary

1. **Hypotheses**

   * $H_0: \mu = 100$ (no effect)
   * $H_1: \mu \neq 100$ (medication has an effect)

2. **Parameters**

   * Sample size: n = 30
   * Sample mean: $\bar{X} = 140$
   * Sample standard deviation: s = 20
   * Degrees of Freedom: $df = n - 1 = 29$
   * Confidence level: 95% → α = 0.05

3. **Critical t-value (two-tailed)**

   * From t-table for df = 29, α = 0.05 → $t_{critical} = ±2.045$

4. **Test Statistic Calculation**

$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}} = \frac{140 - 100}{20 / \sqrt{30}} \approx 10.96
$$

5. **Decision Rule**

   * Reject $H_0$ if $t < -2.045$ or $t > 2.045$

6. **Decision**

   * t\_calculated = 10.96 > 2.045 → **Reject H0**

7. **Conclusion**

   * The medication **has significantly affected intelligence**.
   * Since t > 0, the effect is an **increase in IQ**.

---

**Key Takeaways**

* Use **t-test** when population standard deviation is unknown.
* **Degrees of freedom** affect the t-distribution.
* Compare **t-statistic** with **t-critical values** to decide whether to reject H0.
* Two-tailed test checks for both **increase and decrease** in effect.


## Decision Guide: t-Test vs Z-Test

- Do you know the **population standard deviation (σ)?**
  - **No**
    - Use **t-test**
  - **Yes**
    - Is **sample size n > 30?**
      - **Yes**
        - Use **Z-test**
      - **No**
        - Use **t-test**
