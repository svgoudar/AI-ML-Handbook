
## Performance Metrics

### Mean Absolute Error (MAE)

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

* Measures the **average magnitude of errors** (absolute difference between actual and predicted).
* **Pros**: Simple, interpretable, robust to outliers.
* **Cons**: Doesnâ€™t penalize large errors strongly.

ğŸ‘‰ Example: If house price is \$100,000 and prediction is \$90,000 â†’ error = \$10,000.

---

### Mean Squared Error (MSE)

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

* Measures **average squared error**.
* **Pros**: Penalizes **large errors** more than small ones.
* **Cons**: Not interpretable in the same unit as target (because itâ€™s squared).

ğŸ‘‰ Example: Errors of 10 and 100 â†’ Squared = 100 and 10,000 â†’ large errors dominate.

---

### Root Mean Squared Error (RMSE)

$$
RMSE = \sqrt{MSE}
$$

* Square root of MSE.
* Same units as the target variable â†’ **easier to interpret**.
* Sensitive to outliers (because of squaring).

ğŸ‘‰ If predicting house prices in dollars, RMSE also comes in **dollars**.

---

### R-Squared ( $R^2$ )

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

Where:

* $SS_{res} = \sum (y_i - \hat{y}_i)^2$ â†’ residual sum of squares

* $SS_{tot} = \sum (y_i - \bar{y})^2$ â†’ total variance

* Represents **proportion of variance explained by the model**.

* Values range from **0 to 1** (closer to 1 = better fit).

* Can be negative if model performs worse than baseline mean prediction.

ğŸ‘‰ Example: $R^2 = 0.85$ â†’ model explains **85% of the variance** in the data.

---

### Adjusted R-Squared

$$
R^2_{adj} = 1 - (1 - R^2)\frac{n-1}{n-p-1}
$$

Where:

* $n$ = number of data points

* $p$ = number of features

* Adjusts RÂ² for the **number of predictors**.

* Prevents artificial inflation of RÂ² when irrelevant features are added.

---

### Summary Table

| Metric      | Intuition                 | Penalizes Large Errors? | Units          |
| ----------- | ------------------------- | ----------------------- | -------------- |
| **MAE**     | Avg. absolute error       | âŒ No                    | Same as target |
| **MSE**     | Avg. squared error        | âœ… Yes (strongly)        | Squared units  |
| **RMSE**    | Square root of MSE        | âœ… Yes                   | Same as target |
| **RÂ²**      | Variance explained        | âŒ No                    | Ratio (0â€“1)    |
| **Adj. RÂ²** | RÂ² adjusted for #features | âŒ No                    | Ratio (0â€“1)    |

---

**Rule of Thumb**:

* Use **MAE** when interpretability is important.
* Use **RMSE** when large errors matter more.
* Use **RÂ² / Adjusted RÂ²** to judge overall model fit.
