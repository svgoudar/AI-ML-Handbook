## Probability Density Function (PDF)

The **Probability Density Function (PDF)** describes how probabilities are distributed for **continuous random variables**—variables that can take infinitely many values within a range (e.g., age, weight, salary).

### Key Characteristics of a PDF

* **Distribution of Continuous Variables**
  The PDF shows how probability is spread across possible values. Instead of assigning probability to exact points (as in discrete variables), we calculate the probability over a **range** of values.

* **Properties**

  * **Non-Negativity:** $f(x) \geq 0$ for all $x$.
  * **Total Area = 1:**

    $$
    \int_{-\infty}^{\infty} f(x)\,dx = 1
    $$

    Ensures the random variable takes some value.
  * **Probability as Area:**

    $$
    P(a \leq X \leq b) = \int_a^b f(x)\,dx
    $$

    Probability is represented by the **area under the curve** between $a$ and $b$.

* **Connection to CDF**
  The **Cumulative Distribution Function (CDF)**, $F(x)$, gives the probability that $X \leq x$. The PDF is its derivative:

  $$
  f(x) = \frac{d}{dx}F(x)
  $$

---

### Statistical Measures from PDF

* **Mean (Expected Value):**
  The weighted average of all possible values:

  $$
  \mu = E[X] = \int_{-\infty}^{\infty} x \, f(x)\, dx
  $$

* **Variance:**
  The measure of spread around the mean:

  $$
  \sigma^2 = Var(X) = \int_{-\infty}^{\infty} (x-\mu)^2 f(x)\, dx
  $$

* **Standard Deviation:**
  Square root of variance, gives spread in original units:

  $$
  \sigma = \sqrt{Var(X)}
  $$

* **Median:**
  The value $m$ where half the probability lies below it:

  $$
  P(X \leq m) = 0.5
  $$

* **Mode:**
  The value of $x$ where the PDF reaches its maximum, i.e., the most likely outcome.

---

### Examples of Distributions Using PDF

* **Normal (Gaussian) Distribution** – symmetric bell curve (e.g., heights, test scores).
* **Exponential Distribution** – models waiting times.
* **Log-Normal Distribution** – used for salaries, prices, etc.

---

**In real-world use**: For example, in house price prediction, the **mean** gives the average price, the **variance** shows how spread out prices are, the **median** indicates the typical house price, and the **PDF curve** models the full distribution of house prices.



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for Normal Distribution
mu = 50   # Mean
sigma = 10  # Standard Deviation

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Plot PDF with statistical measures
plt.figure(figsize=(10,6))
plt.plot(x, pdf, label="PDF (Normal Distribution)", color="blue")

# Mean
plt.axvline(mu, color="red", linestyle="--", label=f"Mean = {mu}")

# Median (same as mean for normal)
plt.axvline(mu, color="green", linestyle=":", label=f"Median = {mu}")

# Mode (same as mean for normal)
plt.axvline(mu, color="purple", linestyle="-.", label=f"Mode = {mu}")

# Variance region (1 std dev)
plt.fill_between(x, pdf, where=(x > mu-sigma) & (x < mu+sigma), color="yellow", alpha=0.3, label="1 Std Dev Range")

plt.title("Normal Distribution PDF with Statistical Measures")
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
````
