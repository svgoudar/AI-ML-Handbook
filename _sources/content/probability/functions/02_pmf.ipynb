# Probability Mass Function (PMF):
  * **Used for:** **Discrete random variables** (variables that can only take on specific, distinct values, like whole numbers).
  * **Represents:** The probability of each specific outcome of the discrete random variable.
  * **Example (Rolling a Fair Die):**
      * Outcomes: 1, 2, 3, 4, 5, 6
      * Probability of each outcome: 1/6 (since it's a fair die, each outcome has an equal chance).
      * **PMF Graph:** A bar chart where the x-axis represents the outcomes and the y-axis represents their probabilities. Each bar's height for a fair die would be 1/6.
  * **Key Property:** The sum of all probabilities for all possible outcomes must equal 1.

The **Probability Mass Function (PMF)** is a fundamental concept in probability theory, specifically designed for **discrete random variables**.


* A **PMF** assigns probabilities to the possible values of a **discrete random variable**.
* A discrete random variable can take a **finite** or **countably infinite** set of distinct values.
* Examples include:

  * Number of goals scored in a match
  * Outcome of rolling a die
  * Number of rooms in a house

---

### Example: Rolling a Fair Die

* Suppose a random variable **X** represents the outcome of rolling a die.
* Possible outcomes: {1, 2, 3, 4, 5, 6}.
* Since the die is fair, each outcome has equal probability:

$$
P(X=k) = \frac{1}{6}, \quad k \in \{1,2,3,4,5,6\}
$$

* To calculate a range of probabilities, sum the individual values.

  * Example:

    $$
    P(X \leq 2) = P(X=1) + P(X=2) = \tfrac{1}{6} + \tfrac{1}{6} = \tfrac{2}{6} = \tfrac{1}{3}
    $$

### Common Distributions Using PMF

Several widely used discrete distributions rely on a PMF:

* **Bernoulli Distribution** â†’ Models binary outcomes (success/failure).
* **Binomial Distribution** â†’ Counts number of successes in repeated Bernoulli trials.
* **Poisson Distribution** â†’ Models the number of events occurring in a fixed time or space.
* **Discrete Uniform Distribution** â†’ All outcomes equally likely (e.g., rolling a fair die).

## Statistical Measures

The **PMF** describes the probability distribution of a **discrete random variable**.
For a discrete random variable $X$ taking values $\{x_1, x_2, ..., x_n\}$ with probabilities $p(x_i)$,

$$
p(x_i) = P(X = x_i), \quad \sum_{i} p(x_i) = 1
$$

---

### **Statistical Measures Derived from PMF**

1. **Mean (Expected Value)**
   The mean (expected value) is the **average value** of the distribution:

   $$
   E[X] = \mu = \sum_{i} x_i \cdot p(x_i)
   $$

   *Interpretation*: The long-term average outcome if the random variable is repeated many times.

---

2. **Variance**
   The variance measures the **spread** of the distribution:

   $$
   Var(X) = \sigma^2 = \sum_{i} (x_i - \mu)^2 \cdot p(x_i)
   $$

   Standard Deviation:

   $$
   \sigma = \sqrt{Var(X)}
   $$

---

3. **Mode**
   The mode is the **value of $x_i$** with the highest probability $p(x_i)$:

   $$
   Mode = \arg\max_{x_i} p(x_i)
   $$

---

4. **Median**
   The median is the value $m$ such that:

   $$
   P(X \leq m) \geq 0.5 \quad \text{and} \quad P(X \geq m) \geq 0.5
   $$

   This is found using the **cumulative distribution** derived from the PMF.

---

5. **Higher Moments (Skewness & Kurtosis)**

   * **Skewness**: Measures asymmetry of the distribution.
   * **Kurtosis**: Measures "peakedness" (heavy tails vs light tails).

---

### **Example: Rolling a Fair Die ðŸŽ²**

* Random variable $X$: outcomes = {1,2,3,4,5,6}
* PMF: $p(x) = 1/6$ for each outcome.

**Mean**:

$$
E[X] = \sum_{x=1}^{6} x \cdot \frac{1}{6} = \frac{21}{6} = 3.5
$$

**Variance**:

$$
Var(X) = \sum_{x=1}^{6} (x-3.5)^2 \cdot \frac{1}{6} = 2.92
$$

**Std Dev**:

$$
\sigma = \sqrt{2.92} \approx 1.71
$$

**Mode**: Any of {1,2,3,4,5,6} (all equally likely).

**Median**: Between 3 and 4 (since CDF crosses 0.5 here).



````python
import matplotlib.pyplot as plt
import numpy as np

# Outcomes of a fair die
outcomes = np.arange(1, 7)
probabilities = np.ones(6) / 6  # Equal probability = 1/6

# Plot PMF (bar chart)
plt.bar(outcomes, probabilities, tick_label=outcomes, edgecolor="black")
plt.title("PMF of Rolling a Fair Die")
plt.xlabel("Die Outcome")
plt.ylabel("Probability")
plt.ylim(0, 0.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
