# Hypergeometric Distribution


## ðŸ”¹ **What is the Hypergeometric Distribution?**

The **Hypergeometric Distribution** models the probability of getting a certain number of "successes" when **sampling *without replacement*** from a finite population that contains both successes and failures.

This makes it different from the **Binomial Distribution**, which assumes independent trials (sampling *with replacement* or infinite population).

---

## ðŸ”¹ **Key Parameters**

The hypergeometric distribution depends on 3 main parameters:

1. **Population size (N):** The total number of items in the population.
2. **Number of successes in the population (K):** The count of items labeled as "success" in the population.
3. **Number of draws (n):** How many items are drawn (without replacement).
4. **Random variable (X):** The number of observed successes in those draws.

---

## ðŸ”¹ **Probability Mass Function (PMF)**

$$
P(X = x) = \frac{\binom{K}{x} \cdot \binom{N-K}{n-x}}{\binom{N}{n}}
$$

Where:

* $\binom{K}{x}$ = ways to choose $x$ successes from $K$.
* $\binom{N-K}{n-x}$ = ways to choose the remaining $n-x$ from failures.
* $\binom{N}{n}$ = total ways to choose $n$ items from the population.

---

## ðŸ”¹ **Statistical Measures**

* **Mean (Expected Value):**

  $$
  E[X] = n \cdot \frac{K}{N}
  $$
* **Variance:**

  $$
  Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N-n}{N-1}
  $$

> Notice the extra factor $\frac{N-n}{N-1}$, which accounts for **lack of independence** (since sampling is without replacement).

---

## ðŸ”¹ **Example: Card Drawing**

Suppose you have a **deck of 52 playing cards (N=52)**.

* **Successes (K=4):** Letâ€™s say we define "success" as drawing an Ace.
* **Draws (n=5):** You draw 5 cards without replacement.
* **Question:** Whatâ€™s the probability of getting exactly 2 Aces?

$$
P(X=2) = \frac{\binom{4}{2} \cdot \binom{48}{3}}{\binom{52}{5}}
$$

---

## ðŸ”¹ **Applications**

* Quality control (finding defective items in a batch).
* Card games and lottery problems.
* Ecological studies (sampling species from a habitat).
* Any situation of **finite population sampling without replacement**.

---

âœ… In short:

* Use **Binomial** when trials are independent (replacement allowed or infinite population).
* Use **Hypergeometric** when **sampling without replacement** from a finite group.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Hypergeometric parameters (card example)
N = 52   # population size (deck size)
K = 4    # number of "successes" in population (Aces)
n = 5    # number of draws (cards drawn without replacement)

# Support of X: possible number of aces drawn
x = np.arange(0, n+1)

# PMF
pmf = hypergeom.pmf(x, N, K, n)

# Probability of exactly 2 aces
p_x_eq_2 = hypergeom.pmf(2, N, K, n)

# Mean and variance
mean = hypergeom.mean(N, K, n)
var = hypergeom.var(N, K, n)

# --- Plot PMF ---
plt.figure(figsize=(8,5))
plt.bar(x, pmf)
plt.title(f'Hypergeometric PMF: N={N}, K={K}, n={n}')
plt.xlabel('Number of Aces Drawn (x)')
plt.ylabel('P(X = x)')
for xi, pi in zip(x, pmf):
    plt.text(xi, pi + 0.005, f'{pi:.3f}', ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()

# Print key stats
print(f"P(X = 2) = {p_x_eq_2:.4f}")
print(f"Mean = {mean:.3f}")
print(f"Variance = {var:.3f}")
```
