
## The Poisson Distribution

The **Poisson distribution** is a probability distribution that models the **number of times an event occurs within a fixed interval of time or space**, assuming the events occur independently and at a constant average rate.


### üîë Key Features

* **Purpose**:
  Used to describe and predict how many times an event happens in a given interval.

  * Example: estimating how many users visit a website in one minute, or how many phone calls arrive at a call center in an hour.

* **Parameter**:

  * **$\lambda$ (lambda)** = the expected (mean) number of events per interval.
  * Both the **mean and variance** of a Poisson distribution are equal to $\lambda$.

* **Assumptions**:

  1. Events occur **independently**.
  2. The **rate ($\lambda$) remains constant** within the observed interval.
  3. Two events cannot occur **at exactly the same instant** (events are discrete).

---

#### üìä Characteristics

* A higher $\lambda$ produces a distribution that looks more symmetric and bell-shaped.
* A lower $\lambda$ produces a more skewed distribution with most of the mass near zero.
* Special case in **count data modeling**: Poisson is widely used for event-frequency data, like defects per unit, website clicks, or server requests.

---

#### ‚öôÔ∏è Related Functions in R

* `rpois(n, lambda)`: Generate `n` random values from a Poisson distribution with mean $\lambda$.

  * Example: `rpois(100, lambda = 2)` ‚Üí simulates 100 minutes of calls if the average rate is 2 calls per minute.

---

#### üîó Relationship to Other Distributions

* **Exponential distribution**: Models the **time between consecutive events** when events follow a Poisson process.
* **Weibull distribution**: Generalizes the exponential by allowing event rates to change over time, useful for reliability and survival analysis.
* **Generalized Linear Models (GLMs)**: Poisson is a key distribution for modeling count data in regression.

---

‚úÖ **Key Takeaway**:
The Poisson distribution is ideal for modeling **event counts in fixed intervals**, provided the events occur independently and at a constant average rate.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Parameters
lambdas = [1, 4, 10]   # Different Œª values to demonstrate
x = np.arange(0, 20)   # Range of possible counts

plt.figure(figsize=(12, 8))

# Plot PMFs for different Œª
for i, lam in enumerate(lambdas, 1):
    plt.subplot(2, 3, i)
    pmf = poisson.pmf(x, lam)
    sns.barplot(x=x, y=pmf, color="skyblue", edgecolor="black")
    plt.title(f"Poisson PMF (Œª = {lam})")
    plt.xlabel("Number of events (k)")
    plt.ylabel("P(X = k)")

# Simulated samples for visualization
for i, lam in enumerate(lambdas, 1):
    plt.subplot(2, 3, i+3)
    samples = np.random.poisson(lam, 1000)
    sns.histplot(samples, bins=20, kde=True, color="lightcoral", edgecolor="black")
    plt.title(f"Simulated Data (Œª = {lam})")
    plt.xlabel("Observed counts")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
