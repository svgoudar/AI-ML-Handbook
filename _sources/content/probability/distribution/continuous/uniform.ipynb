# Uniform Continuous Distribution

The **Uniform Continuous Distribution** describes a situation where **all values within a given interval are equally likely** to occur. Unlike the **discrete uniform distribution** (where outcomes are integers like dice rolls), here the variable can take on **any real value within the interval**.

---

## Key Characteristics

1. **Definition**
   A continuous random variable $X$ follows a uniform distribution on the interval $[a, b]$ if:

   $$
   f(x) = 
   \begin{cases} 
   \frac{1}{b-a}, & a \leq x \leq b \\
   0, & \text{otherwise}
   \end{cases}
   $$

   * Here, $a$ is the **minimum value**, and $b$ is the **maximum value**.
   * $f(x)$ is the **probability density function (PDF)**.

2. **Graph Shape**
   The PDF is a **rectangle** between $a$ and $b$, meaning the distribution is **flat**â€”every value is equally probable.

3. **Cumulative Distribution Function (CDF)**
   The CDF is the probability that $X$ takes a value less than or equal to $x$:

   $$
   F(x) = 
   \begin{cases} 
   0, & x < a \\
   \frac{x-a}{b-a}, & a \leq x \leq b \\
   1, & x > b
   \end{cases}
   $$

### Statistical Measures

**Mean (Expected Value):**

$$
E[X] = \frac{a + b}{2}
$$

ðŸ‘‰ The midpoint of the interval.

---

**Variance:**

$$
Var(X) = \frac{(b - a)^2}{12}
$$

ðŸ‘‰ Spread depends only on the interval length.

---

**Standard Deviation:**

$$
\sigma = \frac{b - a}{\sqrt{12}}
$$

---

**Median:**

$$
\text{Median} = \frac{a + b}{2}
$$

ðŸ‘‰ Same as the mean (because of symmetry).

---

**Mode:**
   There is **no single mode**; all values in $[a, b]$ are equally likely.

---

**Skewness:**

$$
\text{Skewness} = 0
$$

ðŸ‘‰ Perfectly symmetric.

---

**Kurtosis (Excess):**

$$
\text{Excess Kurtosis} = -\frac{6}{5} \; \approx -1.2
$$

ðŸ‘‰ Flatter than the normal distribution.


---

## Example

Suppose we pick a random number between $2$ and $8$.

* Here, $a = 2$, $b = 8$.
* PDF: $f(x) = \frac{1}{8-2} = \frac{1}{6}$.
* Mean: $\mu = \frac{2+8}{2} = 5$.
* Variance: $\sigma^2 = \frac{(6)^2}{12} = 3$.
* Probability that $X \leq 5$:

  $$
  F(5) = \frac{5-2}{6} = 0.5
  $$

  â†’ Thereâ€™s a 50% chance the number lies below 5.

---

âœ… Key Idea:
The **uniform continuous distribution** models situations where there is *no bias* within an interval, e.g.:

* A bus arriving anytime between 10 and 10:30 (equally likely).
* Randomly selecting a decimal number between 0 and 1.
* Random time within a fixed duration.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters for the uniform distribution
a, b = 2, 8  # interval [2, 8]

# Create x values
x = np.linspace(a-1, b+1, 1000)

# PDF and CDF
pdf = uniform.pdf(x, loc=a, scale=b-a)
cdf = uniform.cdf(x, loc=a, scale=b-a)

# Plot PDF
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(x, pdf, 'b-', lw=2, label="PDF")
plt.fill_between(x, pdf, alpha=0.3, color='skyblue')
plt.title("Uniform Continuous Distribution (PDF)")
plt.xlabel("x")
plt.ylabel("Density")
plt.axhline(1/(b-a), color="red", linestyle="--", label="Constant Height (1/(b-a))")
plt.legend()

# Plot CDF
plt.subplot(1,2,2)
plt.plot(x, cdf, 'g-', lw=2, label="CDF")
plt.title("Uniform Continuous Distribution (CDF)")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.legend()

plt.tight_layout()
plt.show()
````
