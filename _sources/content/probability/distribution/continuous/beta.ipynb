## Beta Distribution

The **Beta Distribution** is a **continuous probability distribution** defined on the interval **\[0, 1]**.
It is very flexible and is mainly used to model **random variables that represent probabilities or proportions**.


### Parameters

The Beta distribution is controlled by **two shape parameters**:

* **α (alpha)**
* **β (beta)**

The probability density function (PDF) of the Beta distribution is:

$$
f(x; \alpha, \beta) = \frac{1}{B(\alpha, \beta)} x^{\alpha - 1} (1 - x)^{\beta - 1}, \quad 0 < x < 1
$$

Where:

* $B(\alpha, \beta)$ is the **Beta function**, which ensures that the total probability integrates to 1.
* $\alpha > 0, \; \beta > 0$.

---

### Interpretation of Parameters

* **α (alpha)** = prior "successes"
* **β (beta)** = prior "failures"

So, the Beta distribution can be seen as a way of modeling **beliefs about probabilities** based on prior evidence.

---

### Statistical Measures

* **Mean (Expected Value):**

$$
E[X] = \frac{\alpha}{\alpha + \beta}
$$

* **Variance:**

$$
Var[X] = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}
$$

---

### Shapes of the Beta Distribution

Depending on $\alpha$ and $\beta$, the Beta distribution takes different shapes:

* $\alpha = \beta = 1$: **Uniform distribution** (all outcomes equally likely).
* $\alpha > \beta$: Distribution skewed towards 1 (success more likely).
* $\alpha < \beta$: Distribution skewed towards 0 (failure more likely).
* Large $\alpha, \beta$: The distribution becomes more **peaked** (concentrated around the mean).

---

### Applications

1. **Bayesian Statistics** – Beta distribution is the **conjugate prior** for the Bernoulli and Binomial distributions (very common in Bayesian inference).

   * Example: Modeling probability of success (click-through rate, defect rate, etc.).
2. **Proportions and Ratios** – Useful when modeling **fractions** like survival rates, conversion rates, etc.
3. **Random Variables on \[0,1]** – Any variable naturally bounded between 0 and 1 can be modeled with Beta.

---

✅ **In short:**
The Beta distribution is a powerful and flexible way to model **uncertainty in probabilities**. By tuning $\alpha$ and $\beta$, you can represent almost any belief about the likelihood of outcomes between 0 and 1.



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Values for x (between 0 and 1)
x = np.linspace(0, 1, 1000)

# Different parameter settings for demonstration
params = [(1, 1), (2, 5), (5, 2), (2, 2), (10, 10)]

# Plot Beta distributions with different α, β
plt.figure(figsize=(12, 8))

for a, b in params:
    y = beta.pdf(x, a, b)
    plt.plot(x, y, label=f'α={a}, β={b}')

plt.title("Beta Distribution for Different α and β Values", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
````
