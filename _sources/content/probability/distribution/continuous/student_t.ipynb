# Student’s t-Distribution

The **t-distribution** is a probability distribution that is very similar to the **normal distribution** (bell-shaped and symmetric) but has **heavier tails**.

It was introduced by William Sealy Gosset under the pen name *“Student”*.

---

## Key Features

1. **Shape**:

   * Bell-shaped and symmetric around zero.
   * Tails are heavier than the normal distribution → more probability of extreme values.

2. **Parameter**:

   * **Degrees of Freedom (ν or df):**

     * Determines the shape of the distribution.
     * Lower df → heavier tails.
     * As df → ∞, the t-distribution approaches the standard normal distribution.

3. **Probability Density Function (PDF):**

   $$
   f(t) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\sqrt{\nu \pi}\,\Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu+1}{2}}
   $$

   where $\nu$ = degrees of freedom.

### Statistical Measures

**Mean (Expected Value):**

$$
E[X] =
\begin{cases} 
0 & \text{if } \nu > 1 \\
\text{undefined} & \text{if } \nu \leq 1
\end{cases}
$$

The mean is **0** (symmetric about 0) for most cases.

---

**Median:**

$$
Median = 0
$$

Because the distribution is symmetric.

---

**Mode:**

$$
Mode = 0
$$

The peak is always at the center.

---

**Variance:**

$$
Var[X] =
\begin{cases} 
\frac{\nu}{\nu - 2} & \text{if } \nu > 2 \\
\infty & \text{if } 1 < \nu \leq 2 \\
\text{undefined} & \text{if } \nu \leq 1
\end{cases}
$$

So the variance only exists for **ν > 2**.

---

**Standard Deviation:**

$$
SD = \sqrt{\frac{\nu}{\nu - 2}}, \quad \nu > 2
$$

---

**Skewness:**

$$
Skewness = 0, \quad \nu > 3
$$

The distribution is symmetric, so skewness is zero when defined.

---

**Excess Kurtosis:**

$$
Kurtosis = \frac{6}{\nu - 4}, \quad \nu > 4
$$

* For small ν (few degrees of freedom), kurtosis is **large** (heavier tails).
* As ν → ∞, kurtosis → 0 (approaching Normal distribution).

---

## Applications

* **Small Sample Sizes**: Used when the sample size is small and population variance is unknown.
* **t-Tests**: Hypothesis testing for means (e.g., one-sample t-test, two-sample t-test).
* **Confidence Intervals**: Used to calculate confidence intervals for the mean when variance is estimated from the sample.
* **Robustness**: Handles uncertainty better when data is limited.

---

## Relation to Normal Distribution

* Both are symmetric and bell-shaped.
* With large sample sizes (high df), t-distribution ≈ normal distribution.
* With small df, t-distribution has **fatter tails**, making it more conservative.

---

✅ **In short**:
The **Student’s t-distribution** is crucial in **inferential statistics**, especially for **hypothesis testing** and **confidence intervals** when dealing with small samples or unknown population variance.



````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t, norm

# Values
x = np.linspace(-5, 5, 500)

# Degrees of freedom to demonstrate
dfs = [1, 5, 10, 30]

# Plot
plt.figure(figsize=(10, 6))

# Plot t-distributions for different degrees of freedom
for df in dfs:
    plt.plot(x, t.pdf(x, df), label=f"t-distribution (df={df})")

# Plot standard normal distribution for comparison
plt.plot(x, norm.pdf(x, 0, 1), 'k--', label="Normal distribution")

# Style
plt.title("Student's t-distribution vs Normal distribution", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
sns.despine()
plt.grid(alpha=0.3)
plt.show()
````
