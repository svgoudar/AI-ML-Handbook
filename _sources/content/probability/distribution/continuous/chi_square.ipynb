## Chi-Square Distribution

The **Chi-Square Distribution** ($\chi^2$) is a **continuous probability distribution** that arises often in statistics, especially in **hypothesis testing** and **goodness-of-fit tests**.
It is a special case of the **Gamma distribution**.

---

### Parameters

The distribution is determined by a single parameter:

* **Degrees of freedom (k or df)** – a positive integer that usually corresponds to the number of independent standard normal random variables being squared and summed.

If $Z_1, Z_2, \dots, Z_k \sim N(0,1)$, then

$$
X = \sum_{i=1}^k Z_i^2 \sim \chi^2(k)
$$

---

### Probability Density Function (PDF)

For a Chi-square random variable $X \sim \chi^2(k)$:

$$
f(x; k) = \frac{1}{2^{k/2}\Gamma(k/2)} x^{\frac{k}{2}-1} e^{-x/2}, \quad x > 0
$$

Where:

* $k$ = degrees of freedom
* $\Gamma(\cdot)$ = Gamma function

---

### Statistical Measures

* **Mean (Expected Value):**

$$
E[X] = k
$$

* **Variance:**

$$
Var[X] = 2k
$$

* **Standard Deviation:**

$$
\sigma = \sqrt{2k}
$$

* **Mode:**

$$
\text{Mode} =
\begin{cases}
k - 2 & \text{if } k \geq 2 \\
0 & \text{if } 0 < k < 2
\end{cases}
$$

* **Skewness:**

$$
\text{Skewness} = \sqrt{\frac{8}{k}}
$$

* **Kurtosis (Excess):**

$$
\text{Kurtosis} = \frac{12}{k}
$$

---

### Shapes of the Chi-Square Distribution

* **Small df (1–2):** Highly skewed right, peak near 0.
* **Moderate df (5–10):** Skew decreases, distribution spreads.
* **Large df (>30):** Approaches a **Normal distribution** $N(k, 2k)$.

---

### Applications

1. **Goodness-of-Fit Tests** – Compare observed vs. expected frequencies.
2. **Test of Independence** – Used in contingency tables (Chi-square test of independence).
3. **Variance Estimation** – Sampling distribution of variance is Chi-square.
4. **Confidence Intervals** – For population variance and standard deviation.

---

✅ **In short:**
The Chi-square distribution is central to **inferential statistics**, especially for tests involving **categorical data, variances, and independence**. Its shape depends on the **degrees of freedom**, and with large df, it approximates the **normal distribution**.



````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2

# Values for x-axis
x = np.linspace(0, 30, 500)

# Different degrees of freedom
dfs = [1, 2, 5, 10, 20]

plt.figure(figsize=(10, 6))

# Plot Chi-Square distributions
for df in dfs:
    plt.plot(x, chi2.pdf(x, df), label=f"df={df}")

# Plot styling
plt.title("Chi-Square Distribution for Different Degrees of Freedom", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
````
