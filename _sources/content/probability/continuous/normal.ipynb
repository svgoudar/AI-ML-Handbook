## Normal (Gaussian) Distribution

### 1. Definition

The **normal distribution**, also called the **Gaussian distribution**, is the most common probability distribution in statistics.
It is used to model variables that tend to **cluster around a central value** with symmetrical variation.

Mathematically, the probability density function (PDF) is:

$$
f(x | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Where:

* **Î¼ (mu)** â†’ mean (center of the curve)
* **Ïƒ (sigma)** â†’ standard deviation (spread/width of the curve)
* **ÏƒÂ²** â†’ variance

---

### 2. Key Properties

* **Symmetry** â†’ perfectly symmetric about the mean Î¼.
* **Mean = Median = Mode** (all are equal in a normal distribution).
* **68-95-99.7 Rule (Empirical Rule)**

  * \~68% of values lie within **1Ïƒ** of Î¼
  * \~95% of values lie within **2Ïƒ** of Î¼
  * \~99.7% of values lie within **3Ïƒ** of Î¼
* **Total Area = 1** â†’ probability distribution requirement.

---

### 3. Applications

* **Heights, weights, IQ scores, exam results** often approximate normal distribution.
* **Error terms** in regression analysis often assumed normal.
* Used in **Central Limit Theorem (CLT)**: averages of samples tend to become normally distributed.

---

### 4. Intuition with Bell Curve

Imagine peopleâ€™s heights:

* Most people are around the average (Î¼).
* Very tall/very short people are rare â†’ values taper off symmetrically on both sides.

ðŸ“Š This creates the **bell-shaped curve** (Gaussian curve).



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu, sigma = 0, 1  # mean and standard deviation

# Generate data points
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Plot PDF (bell curve)
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label="PDF (Probability Density Function)", color="blue")
plt.fill_between(x, pdf, alpha=0.2, color="blue")

# Mark mean and std deviations
for k, perc in zip([1, 2, 3], [0.68, 0.95, 0.997]):
    plt.axvline(mu + k*sigma, color="red", linestyle="--", alpha=0.7)
    plt.axvline(mu - k*sigma, color="red", linestyle="--", alpha=0.7)
    plt.text(mu + k*sigma, 0.02, f"{perc*100:.1f}%", color="red", ha="left")

plt.axvline(mu, color="black", linestyle="-", label="Mean (Î¼)")

# Labels and title
plt.title("Normal (Gaussian) Distribution\nMean = 0, Std Dev = 1")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
